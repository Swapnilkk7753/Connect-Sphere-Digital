```python
dff2=pd.concat([dff,dff1],ignore_index=True)
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
x=dff2[['Daily Time Spent on Site','Age', 'Area Income','Daily Internet Usage']]
y=dff2['Clicked on Ad']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=LogisticRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
```

```python
y_pre=model.predict(x_test)
print("Accuracy Score->",accuracy_score(y_test,y_pre))
print("Precision Score->",precision_score(y_test,y_pre))
print("Recall Score->",recall_score(y_test,y_pre))
print("F1 Score->",precision_score(y_test,y_pre))
print("Confusion Matrix->\n",confusion_matrix(y_test,y_pre))
```
